```python
from unsloth import FastLanguageModel
import torch

# Configuration parameters
max_seq_length = 2048  # You can adjust this as needed
dtype = None            # None for auto-detection. Set to torch.float16 for specific GPUs if needed
load_in_4bit = True     # Enables 4-bit quantization to reduce memory usage

# Load the pretrained model and tokenizer
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/llama-3-8b-bnb-4bit",  # Replace with your desired model
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
    # token="hf_...",  # Uncomment and set your Hugging Face token if using gated models
)
```

```python
from datasets import load_dataset
from unsloth import to_sharegpt, standardize_sharegpt, apply_chat_template

# Load the dataset
dataset = load_dataset("vicgalle/alpaca-gpt4", split="train")
print("Dataset Column Names:", dataset.column_names)

# Transform the dataset to ShareGPT format
dataset = to_sharegpt(
    dataset,
    merged_prompt="{instruction}[[\nYour input is:\n{input}]]",
    output_column_name="output",
    conversation_extension=3,  # Increase if handling longer conversations
)

# Standardize the dataset format
dataset = standardize_sharegpt(dataset)

# Define a custom chat template compatible with OpenAI-style chatbots
chat_template = """Below are some instructions that describe some tasks. Write responses that appropriately complete each request.

### Instruction:
{INPUT}

### Response:
{OUTPUT}"""

# Apply the chat template to the dataset
dataset = apply_chat_template(
    dataset,
    tokenizer=tokenizer,
    chat_template=chat_template,
    # default_system_message="You are a helpful assistant",  # Optional
)

print("Dataset has been preprocessed and is ready for training.")
```

```python
from unsloth import FastLanguageModel

# Attach LoRA adapters to the model
model = FastLanguageModel.get_peft_model(
    model,
    r=16,  # Rank for LoRA. Higher values can capture more complex adaptations.
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha=16,
    lora_dropout=0,          # Dropout rate for LoRA layers. 0 is optimized.
    bias="none",             # Bias configuration. "none" is optimized.
    use_gradient_checkpointing="unsloth",  # Enables gradient checkpointing for reduced memory usage.
    random_state=3407,       # Seed for reproducibility.
    use_rslora=False,        # Whether to use rank stabilized LoRA.
    loftq_config=None,       # Configuration for LoftQ if used.
)

print("LoRA adapters have been successfully attached to the model.")
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported
import torch

# Initialize the trainer with LoRA adapters
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",  # The field in the dataset containing the training text
    max_seq_length=max_seq_length,
    dataset_num_proc=2,          # Number of processes for data loading
    packing=False,                # Set to True if you want to pack multiple sequences into a single example
    args=TrainingArguments(
        per_device_train_batch_size=2,          # Batch size per GPU
        gradient_accumulation_steps=4,          # Accumulate gradients over 4 steps
        warmup_steps=5,                         # Number of warmup steps for learning rate scheduler
        max_steps=60,                            # Total number of training steps
        # num_train_epochs=1,                    # Uncomment for training based on epochs instead of steps
        learning_rate=2e-4,                      # Learning rate
        fp16=not is_bfloat16_supported(),        # Use FP16 if BFLOAT16 is not supported
        bf16=is_bfloat16_supported(),            # Use BFLOAT16 if supported
        logging_steps=1,                         # Log training progress every step
        optim="adamw_8bit",                      # Optimizer
        weight_decay=0.01,                       # Weight decay for regularization
        lr_scheduler_type="linear",              # Learning rate scheduler
        seed=3407,                               # Random seed for reproducibility
        output_dir="outputs",                    # Directory to save training outputs
        report_to=[],                             # Disable WandB and other reporting integrations
        # You can also explicitly set `logging_dir` if you want to store logs locally
        # logging_dir="logs",
    ),
)

# Display current GPU memory usage before training
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024**3, 3)
max_memory = round(gpu_stats.total_memory / 1024**3, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")

# Start training
trainer_stats = trainer.train()

# Display final GPU memory usage and training time
used_memory = round(torch.cuda.max_memory_reserved() / 1024**3, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round((used_memory / max_memory) * 100, 3)
lora_percentage = round((used_memory_for_lora / max_memory) * 100, 3)

print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(f"{round(trainer_stats.metrics['train_runtime'] / 60, 2)} minutes used for training.")
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage}%.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage}%.")
```

```python
# Save the LoRA adapters locally
model.save_pretrained("lora_model")  # Local saving
tokenizer.save_pretrained("lora_model")

print("LoRA adapters and tokenizer have been saved to the 'lora_model' directory.")
```

```python
# Export the fine-tuned model to GGUF format
model.save_pretrained_gguf(
    "model",
    tokenizer=tokenizer,
    quantization_method="q8_0"  # Options: "q8_0", "q4_k_m", "q5_k_m"
)

print("Model has been successfully exported to GGUF format in the 'model' directory.")
```
